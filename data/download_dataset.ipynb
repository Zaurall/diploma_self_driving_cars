```python
import os
import requests
import zipfile
import shutil
from huggingface_hub import HfApi, hf_hub_download, login
```

```python
# %env HF_TOKEN=YOUR_HF_TOKEN
```

```python
REPO_ID = "commaai/commaSteeringControl"
DATA_DIR = "data"
LOCAL_DIR = "train"

os.makedirs(LOCAL_DIR, exist_ok=True)

api = HfApi()

token = os.getenv("HF_TOKEN")

login(token)
```

```python
files = api.list_repo_files(REPO_ID, repo_type="dataset")

csv_files = [f for f in files if f.startswith(DATA_DIR) and (f.endswith(".csv") or f.endswith(".zip"))]

print(f"Found {len(csv_files)} CSV/ZIP files to download.")
```

```python
for f in csv_files:
    print(f"Downloading {f} ...")
    local_path = hf_hub_download(repo_id=REPO_ID, filename=f, repo_type="dataset", local_dir=LOCAL_DIR)

    # If it's a zip, extract it
    if local_path.endswith(".zip"):
        with zipfile.ZipFile(local_path, "r") as zip_ref:
            zip_ref.extractall(LOCAL_DIR)
        print(f"Extracted {local_path}")

print("✅ All files downloaded and extracted.")
```

```python
shutil.rmtree(os.path.join(LOCAL_DIR, DATA_DIR))
```
